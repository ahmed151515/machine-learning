```python
import pandas as pd

data = pd.read_csv("student-mat.csv")
```

traveltime,studytime,failures,freetime
schoolsup,famsup,paid,higher,internet

```python
columns_to_convert  = ["schoolsup","famsup","paid","higher","internet"]

data[columns_to_convert] = data[columns_to_convert].replace({"yes":1, "no":0})
data[columns_to_convert] = data[columns_to_convert].astype(int)
pd.set_option('display.max_columns', 33)
data.head()
```

```python
from sklearn.model_selection import train_test_split

x = data[["schoolsup","famsup","paid","higher","internet",
"traveltime","studytime","failures","freetime"
]]
y = data[["G1","G2","G3"]]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, random_state=33)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

y_pred =model.predict(x_test)


mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R² Score: {r2}')
print(f"Score: {model.score(x_test, y_test)}")
```
